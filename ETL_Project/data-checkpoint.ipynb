```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/LA.csv"
LA_df = pd.read_csv(csv_file)
LA_df.head()
```

```python
new_LA_data_df = LA_df[['facility_name','facility_city', 'facility_state','score','activity_date' ]].copy()
new_LA_data_df.head()
```

```python
new_LA_data_df.describe()
```

```python
new_LA_data_df=new_LA_data_df.rename(columns = {'facility_name':'business','facility_city':'city',\
                            'facility_state':'state','score':'score','activity_date':'date'})
new_LA_data_df.head()
```

```python
csv_file = "Resources/SF.csv"
SF_df = pd.read_csv(csv_file)
SF_df.head()
```

```python
new_SF_df = SF_df[["business_name", "business_city", "business_state", "inspection_score","inspection_date"]].copy()
new_SF_df.head()
```

```python
new_SF_df.describe()
```

```python
new_SF_df=new_SF_df.rename(columns = {'business_name':'business','business_city':'city',\
                            'business_state':'state','inspection_score':'score','inspection_date':'date'})
new_SF_df.head()
```

```python
frames = [new_SF_df, new_LA_data_df]

combined_df = pd.concat(frames)

combined_df.head(100)
```

```python
combined_df.describe()
```

```python
combined_df.isna().sum()
```

```python
combined_df=combined_df.dropna(how='any')
```

```python
combined_df['date']=pd.to_datetime(combined_df['date'])
```
